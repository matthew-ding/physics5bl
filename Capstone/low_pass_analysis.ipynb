```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def Vp_to_dBm(x):
    return mW_to_dBm(Vp_to_mW(x)) # Vp -> mW -> dBm

def Vp_to_mW(x):
    return (x ** 2 / 2) * (1e3 / 50) # Vp is peak voltage, Vrms^2 = Vp^2/2

def mW_to_dBm(x):
    return 10. * np.log10(x)
```

```python
dataframe = pd.read_csv('data.csv')
dataframe
```

```python
### TRIAL 1 ###
frequency = pd.DataFrame.to_numpy(dataframe)[1:,0]
frequency = np.array([float(x) for x in frequency])
frequency
voltage = pd.DataFrame.to_numpy(dataframe)[1:,1]
voltage = np.array([float(x) for x in voltage])
voltage
```

```python
plt.plot(frequency, voltage)
```

```python
voltage = Vp_to_dBm(voltage)
frequency = np.log(frequency)
plt.plot(frequency, voltage)
```
