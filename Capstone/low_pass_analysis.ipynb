```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def Vp_to_dBm(x):
    return mW_to_dBm(Vp_to_mW(x)) # Vp -> mW -> dBm

def Vp_to_mW(x):
    return (x ** 2 / 2) * (1e3 / 50) # Vp is peak voltage, Vrms^2 = Vp^2/2

def mW_to_dBm(x):
    return 10. * np.log10(x)
```

```python
dataframe = pd.read_csv('data.csv')
dataframe
```

```python
### TRIAL 1 ###
frequency1 = pd.DataFrame.to_numpy(dataframe)[1:,0]
frequency1 = np.array([float(x) for x in frequency1])
frequency1
voltage1 = pd.DataFrame.to_numpy(dataframe)[1:,1]
voltage1 = np.array([float(x) for x in voltage1])
voltage1
```

```python
plt.title('Trial 1 (Raw Data)')
plt.xlabel('Frequency (hZ)')
plt.ylabel('Voltage (dB)')
plt.plot(frequency1, voltage1)
```

```python
voltage1 = Vp_to_dBm(voltage1)
frequency1 = np.log(frequency1)
plt.title('Trial 1')
plt.xlabel('Log Frequency (log hZ)')
plt.ylabel('Voltage (dB)')
plt.plot(frequency1, voltage1)
```

```python
### TRIAL 2 ###
frequency2 = pd.DataFrame.to_numpy(dataframe)[1:,4]
frequency2 = np.array([float(x) for x in frequency2])
frequency2
voltage2 = pd.DataFrame.to_numpy(dataframe)[1:,5]
voltage2 = np.array([float(x) for x in voltage2])
voltage2
voltage2 = Vp_to_dBm(voltage2)
frequency2 = np.log(frequency2)
plt.title('Trial 2')
plt.xlabel('Log Frequency (log hZ)')
plt.ylabel('Voltage (dB)')
plt.plot(frequency2, voltage2, 'pink')
```
