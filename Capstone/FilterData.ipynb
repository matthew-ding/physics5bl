```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def Vp_to_dBm(x):
    return mW_to_dBm(Vp_to_mW(x)) # Vp -> mW -> dBm

def Vp_to_mW(x):
    return (x ** 2 / 2) * (1e3 / 50) # Vp is peak voltage, Vrms^2 = Vp^2/2

def mW_to_dBm(x):
    return 10. * np.log10(x)

def Vout_to_gain(Vout, Vin):
    return 20 * np.log10(Vout / Vin)
```
